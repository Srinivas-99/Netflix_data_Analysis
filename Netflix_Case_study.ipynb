### **Data cleaning**

**Handling Missing Data**

```python
netflix_df.isna().sum()
```

```python
# Checking the percentage of the nulls in the each column
for i in netflix_df.columns:
  null_rate= (netflix_df[i].isna().sum())/len(netflix_df[i]) *100
  if (null_rate>0):
    print(f"{i}'s null percentage rate : {round(null_rate,2)}%\n")
```

Based on the above information there are more null values in the director, cast and country columns compared to the date_added,rating and duration

```python
# - updating the null values with the data
data=netflix_df.copy()

# - for director, cast and country columns updating the null values with NA
data.loc[:,'director']=data.loc[:,'director'].fillna('NA')
data.loc[:,'cast']=data.loc[:,'cast'].fillna('NA')
data.loc[:,'country']=data.loc[:,'country'].fillna('NA')
```

```python
# - Filling missing values with mode for columns 'date_added', 'rating' and duration.
data['date_added'].fillna(data['date_added'].mode()[0],inplace=True)
data.rating.fillna(data.rating.mode()[0], inplace=True)
```

```python
data.rating.value_counts(dropna=False)
```

we can see the durations are added into the rating columns so moving them to the duration column

```python
data.loc[data['rating'].str.contains('min'),'duration']=data.loc[data['rating'].str.contains('min'),'rating']
# - Replacing the duration values in the rating column with mode values
data.loc[data['rating'].str.contains('min'),'rating'] = data['rating'].mode()[0]
```

```python
data.isna().sum()
```

```python
# - Converting 'date_added' column to datetime and adding new columns.

data['date_added'] = pd.to_datetime(data['date_added'])
data['year_added'] = data['date_added'].dt.year
data['month_added'] = data['date_added'].dt.month_name()
data['day_added'] = data['date_added'].dt.day_name()
```

Converting the fields into category type

```python
# - Converting apropriate columns to category type.

data = data.astype({
    'type':'category',
    'rating':'category'
})
```

```python
data.info()
```

We can that now the data is clean and there are no null values in the dataset


**Handling Nested data:**

In the data there are many columns with nested data. so, if we unnest all at a time the dataset will get more duplicate records and the analysis will go wrong.
so Created the below function that will unnest the data when ever it need based on a column.


```python
# - Function to unnest data

def unnest_data(df, col : str, name : str, along:str = 'year_added'):
  segments= data[col].apply(lambda x: str(x).split(', ')).to_list()
  df_new=pd.DataFrame(segments,index=data['title'])
  df_new=df_new.stack()
  df_country= pd.DataFrame(df_new)
  df_country.reset_index(inplace=True)
  df_country=df_country[['title',0]]
  df_country.rename(columns={0:name},inplace=True)
  df_country.replace('NA',np.nan,inplace=True)
  df_country.dropna(inplace=True)
  df_c=data.loc[:,[along,'title']].merge(df_country,left_on='title',right_on='title',how='right')
  df=df_c.drop("title",axis=1)
  #df[name].dropna()
  return df
```
